```python

def getbhk(s):
    n=s.split()[0]
    return int(n)
df3['bhk'] = df3['size'].apply(lambda x:getbhk(x))
```

```python
df4=df3.drop(['size'] ,axis=1)
```

```python
def isfloat(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df4[~df4['total_sqft'].apply(isfloat)].total_sqft.unique()
```

```python
def conv_sqft(x):
    s=x.split('-')
    if(len(s)==2):
        return float(float(s[0])+float(s[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df4['area'] = df4['total_sqft'].apply(conv_sqft)
df5=df4.drop(['total_sqft'],axis=1)
```

```python
df5['price/sq.feet'] = df5['price']*100000/df5['area']
df5.head()
```

```python
df5['location'] = df5['location'].apply(lambda x: x.strip())
stats=df5.groupby('location')['location'].agg('count').sort_values(ascending=False)
```

```python
df5.location=df5['location'].apply(lambda x:'other' if x in stats_less_ten else x.strip())
```

```python
df6 = df5[~(df5.area/df5.bhk < 300)]
```

```python
df6.shape
```

```python
def remove_pps_outlier(df):
    df_out=pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m=np.mean(subdf['price/sq.feet'])
        st=np.std(subdf['price/sq.feet'])
        reduced_df  = subdf[(subdf['price/sq.feet'] > (m-st)) & (subdf['price/sq.feet'] <= (m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
    
```

```python
df7=remove_pps_outlier(df6)
```

```python
len(df7.location.unique())
```

```python
df7.shape
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats={}
        for bhk,bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df['price/sq.feet']),
                'std': np.std(bhk_df['price/sq.feet']),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices,bhk_df[bhk_df['price/sq.feet']<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
```

```python
df8 = remove_bhk_outliers(df7)
df8.shape
```

```python
df9 = df8.drop(['price/sq.feet'],axis=1)
```

```python
df9=df9[df9.bath <df9.bhk+2]
```

```python
df9.shape
```

```python
dummies = pd.get_dummies(df9.location)
```

```python
dummies
```

```python
df10 = pd.concat([df9,dummies.drop(['other'],axis = 1)],axis=1)
df10
```

```python
df11=df10.drop('location',axis=1)
df11
```

```python
df11.shape
```

```python
X= df11.drop('price',axis=1)
y=df11['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=10)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_mdl = LinearRegression()
```

```python
lr_mdl.fit(X_train,y_train)
```

```python
lr_mdl.score(X_test,y_test)
```

```python
def predict_price(location,bath,bhk,area):
    loc = np.where(X.columns==location)[0][0]

    x=np.zeros(len(X.columns))
    x[0] = bath
    x[1] = bhk
    x[2] = area
    if(loc >= 0):
        x[loc]=1
    return lr_mdl.predict([x])[0]
```

```python
predict_price('1st Phase JP Nagar',2,2,1000)
```

```python
predict_price('1st Phase JP Nagar',3,3,1000).round(0)
```
